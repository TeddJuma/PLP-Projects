# __Task 2: Basic Data Analysis__

```python
print(df.describe())

unique_species = df["Species"].unique()
print(f"\nThese are the unique species present in the dataset: \n{unique_species} \nThere are {len(unique_species)} unique species. \n")

print(df["Species"].value_counts())
```

### NOTES:
* The Iris dataset contains 150 observations with 5 variables: Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, and PetalWidthCm.
* The mean values for the sepal and petal measurements are: SepalLengthCm (5.84 cm), SepalWidthCm (3.05 cm), PetalLengthCm (3.76 cm), and PetalWidthCm (1.20 cm).
* The sepal length and width measurements have a relatively small range, while the petal length and width measurements have a larger range.
* The mean values for the petal measurements are larger than the mean values for the sepal measurements.
* The data shows a positive correlation between sepal length and width, and a roughly normal distribution for sepal length.
* There are 3 species of the Iris flower present in the dataset.

# Task 3: Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

# A Line graph of the Sepal Length for each species
plt.plot(df["SepalLengthCm"])
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.title("Sepal Length over Index")
plt.show()

# A Bar Plot of the average Sepal Length for each species
avg_sepal_length = df.groupby("Species")["SepalLengthCm"].mean()
plt.bar(avg_sepal_length.index, avg_sepal_length.values)
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.title("Average Sepal Length by Species")
plt.show()

# A Histogram of the Petal Length for each species
plt.hist(df["PetalLengthCm"], bins=10, color="purple")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Petal Length Distribution")
plt.show()

# A scatter plot showing the relationship between Sepal Length and Petal Length, colored by species
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"], c=df["Species"].astype("category").cat.codes, cmap="viridis")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```
